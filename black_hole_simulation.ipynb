```python
import matplotlib.pyplot as plt
import numpy as np
G = 6.67430e-11
c = 299792458
```

```python
def schwarzschild_radius(M):
  return 2*G*M/c**2
```

```python
def redshift_factor(r, rs):
  return (1-rs/r)**(1/2)
```

```python
def received_frequency(f_emit, r, rs):
  return f_emit*redshift_factor(r, rs)
```

```python
def received_wavelength(lambda_emit, r, rs):
  return lambda_emit/(redshift_factor(r, rs))
```

```python
M=10**40
f_emit=200
rs=schwarzschild_radius(M)

r_list = [1.00000001*rs, 1.0001*rs, 1.05*rs, 1.1*rs, 1.5*rs, 2.0*rs, 3.0*rs, 5.0*rs]
received_frequencies=[]
for r in r_list:
  received_frequencies.append(received_frequency(f_emit, r, rs))

plt.plot(r_list, received_frequencies)
plt.xlabel('r')
plt.ylabel('Received Frequency')
plt.axhline(y=20, color='g', linestyle='--')
# below green line ppl cannot hear
```

```python
# radiowave @ set M=10**40
M = 10**40
f_emit = 200 # (VLF radio wave frequency)

rs = schwarzschild_radius(M)

r_list = np.array([1.00000001*rs, 1.0001*rs, 1.05*rs, 1.1*rs, 1.5*rs, 2.0*rs, 3.0*rs, 5.0*rs, 10.0*rs, 50.0*rs, 100.0*rs])

received_frequencies = []
for r_val in r_list:
  received_frequencies.append(received_frequency(f_emit, r_val, rs))

plt.figure(figsize=(10, 6))
plt.plot(r_list / rs, received_frequencies, marker='o') # Plotting r in units of rs for better scale
plt.xlabel('Distance from Black Hole (r/rs)')
plt.ylabel('Received Frequency (Hz)')
plt.title('Received Radiowave Frequency vs. Distance from Black Hole (M=10^40 kg)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(y=f_emit, color='red', linestyle=':', label='Emitted Frequency')
plt.legend()
plt.show()
```

```python
r_fixed = 5e9 # meters
f_emit = 200

M_list = np.logspace(30, np.log10(3e36), 100) # From 10^30 kg to 3*10^36 kg, 100 points
# defines a range of black hole masses

received_frequencies = []
valid_M_list = []

for M_val in M_list:
  rs_val = schwarzschild_radius(M_val)
  if r_fixed > rs_val:
    freq = received_frequency(f_emit, r_fixed, rs_val)
    received_frequencies.append(freq)
    valid_M_list.append(M_val)

plt.figure()
plt.plot(valid_M_list, received_frequencies)
plt.xlabel('black hole mass (kg)')
plt.ylabel('received frequency (Hz)')
```

```python
# altering sound with varying masses
import numpy as np
import soundfile as sf
import librosa

audio, sr = sf.read("i lawb kitty cats.wav")

# change mass for goofy effects^tm
M = 3e36
r_fixed = 5e9

rs_val = schwarzschild_radius(M)

# make sure that r_fixed is greater than rs_val to avoid issues with redshift_factor
if r_fixed > rs_val:
  alpha = redshift_factor(r_fixed, rs_val)
  audio_shifted = librosa.resample(audio, orig_sr=sr, target_sr=int(sr / alpha))

  sf.write("receivedM3e36.wav", audio_shifted, int(sr))
```
